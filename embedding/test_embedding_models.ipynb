```python
# Model definitions
class HuggingFaceSentenceTransformer:
    def __init__(self, model_name, model_name_short, **kwargs):
        self.model = SentenceTransformer(model_name, **kwargs)
        self.model_name_short = model_name_short

    def embed(self, text):
        return self.model.encode(text)

class NomicEmbedTextV1(HuggingFaceSentenceTransformer):
    def __init__(self):
        super().__init__("nomic-ai/nomic-embed-text-v1", "NomicEmbedTextV1", trust_remote_code=True)

class Robbert2022Dutch(HuggingFaceSentenceTransformer):
    def __init__(self):
        super().__init__('NetherlandsForensicInstitute/robbert-2022-dutch-sentence-transformers', "Robbert2022Dutch")

class BioLord2023(HuggingFaceSentenceTransformer):
    def __init__(self):
        super().__init__('FremyCompany/BioLORD-2023-M-Dutch-InContext-v1', "BioLORD2023")

class AllnlniGroNLP(HuggingFaceSentenceTransformer):
    def __init__(self):
        super().__init__('textgain/allnli-GroNLP-bert-base-dutch-cased', "Allnli")

class Robbert2022DutchFinetuned(HuggingFaceSentenceTransformer):
    def __init__(self):
        super().__init__("Robbert2022DutchFinetuned", "Robbert2022DutchFinetuned")
```

```python
def create_similarity_tuples(data):
    question = data['question']
    relevant_texts = data['relevant_texts']
    irrelevant_texts = data['irrelevant_texts']
    
    similar_texts = []
    dissimilar_texts = []
    
    # Similar texts: question with each relevant text
    similar_texts.extend([(question, text) for text in relevant_texts])
    
    # Similar texts: all combinations of relevant texts
    similar_texts.extend(list(combinations(relevant_texts, 2)))
    
    # Dissimilar texts: question with each irrelevant text
    dissimilar_texts.extend([(question, text) for text in irrelevant_texts])
    
    return similar_texts, dissimilar_texts

# Initialize empty lists to store all similar and dissimilar texts
all_similar_texts = []
all_dissimilar_texts = []

# Iterate over all entries in test_cases
for test_case in test_cases:
    similar_texts, dissimilar_texts = create_similarity_tuples(test_case)
    all_similar_texts.extend(similar_texts)
    all_dissimilar_texts.extend(dissimilar_texts)

# Now all_similar_texts and all_dissimilar_texts contain the combined results from all test cases
```
